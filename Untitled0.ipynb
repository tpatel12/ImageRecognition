```python
from skimage import io
img = io.imread('yale.png') 
io.imshow(img)
```

```python
from skimage.color import rgb2gray 
grey_img = rgb2gray(img) 
 
io.imshow(grey_img) 
io.show() 
```

```python
from skimage.filters import sobel 
sobel_edge = sobel(grey_img) 
io.imshow(sobel_edge)
io.show()

```

```python
from skimage.filters import roberts 
robert_cross_edge = roberts(grey_img) 
io.imshow(robert_cross_edge)
io.show()
```

```python
from skimage import feature 
import matplotlib.pyplot as plt 

canny_edge = feature.canny(grey_img, sigma = 2.5, low_threshold = 0.1 , high_threshold = 0.5)

plt.imshow(canny_edge) 
plt.show() 

print(canny_edge)
```
